```python
import pandas as pd

# Load the dataset from the CSV file
df = pd.read_csv('/content/Q3_Student_Dataset_DT_KNN_Noise.csv')

# Display all rows of the DataFrame
display(df)
```

```python
from sklearn.preprocessing import StandardScaler

# Identify numerical input features for scaling
features_to_scale = ['Attendance_Percent', 'Assignment_Avg', 'Quiz_Avg']

# Initialize the StandardScaler
scaler = StandardScaler()

# Apply standardization to the selected features
df[features_to_scale] = scaler.fit_transform(df[features_to_scale])

print("DataFrame after Feature Scaling (Standardization):")
display(df.head())
```

### 1. Prepare the Data (Features and Target)

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Separate features (X) and target (y)
X = df[['Attendance_Percent', 'Assignment_Avg', 'Quiz_Avg']]
y = df['Final_Result']

# Encode the target variable 'Final_Result' (Pass/Fail) into numerical (1/0)
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print("Original target classes:", le.classes_)
print("Encoded target values (0 for 'Fail', 1 for 'Pass'):", y_encoded[:5])
```

### 2. Split the Data into Training and Testing Sets

```python
# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

print(f"Training features shape: {X_train.shape}")
print(f"Testing features shape: {X_test.shape}")
print(f"Training target shape: {y_train.shape}")
print(f"Testing target shape: {y_test.shape}")
```

### 3. Train the Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize the Decision Tree Classifier
dt_model = DecisionTreeClassifier(random_state=42)

# Train the model
dt_model.fit(X_train, y_train)

print("Decision Tree model trained successfully.")
```
