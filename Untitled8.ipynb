```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with missing values
data = {
    'attendance': [90, 85, np.nan, 70, 95, 88],
    'assignment': [75, np.nan, 80, 65, 90, 78],
    'quizzes': [80, 70, 85, np.nan, 92, 75],
    'final_result': ['Pass', 'Fail', 'Pass', 'Fail', 'Pass', 'Pass']
}
df = pd.DataFrame(data)

print("Original DataFrame with missing values:")
display(df)
print("\nMissing values count per column:")
display(df.isnull().sum())
```

### Handling Missing Values: Imputation with Mean

This method replaces missing numerical values with the mean of their respective columns. This is a good strategy when the missing data is not extensive and the mean is a representative value.

```python
# Impute missing numerical values with the mean
df_imputed_mean = df.copy()
for col in ['attendance', 'assignment', 'quizzes']:
    if df_imputed_mean[col].isnull().any():
        mean_value = df_imputed_mean[col].mean()
        df_imputed_mean[col].fillna(mean_value, inplace=True)

print("DataFrame after imputing missing values with the mean:")
display(df_imputed_mean)
print("\nMissing values count per column after mean imputation:")
display(df_imputed_mean.isnull().sum())
```

### Handling Missing Values: Deleting Rows

This method removes rows that contain any missing values. Use this with caution, as it can lead to significant data loss if many rows have missing data. It's often suitable when missing values are sparse and you want to ensure a complete record for each entry.

```python
# Drop rows with any missing values
df_dropped = df.copy()
df_dropped.dropna(inplace=True)

print("DataFrame after dropping rows with missing values:")
display(df_dropped)
print("\nMissing values count per column after dropping rows:")
display(df_dropped.isnull().sum())
```

Choose the method that best suits your dataset and the impact of missing data. For your actual dataset, you would replace the `df = pd.DataFrame(data)` line with your data loading code (e.g., `df = pd.read_csv('your_data.csv')`).

### Handling Missing Values: Feature Scaling - Standardization (Z-score normalization)

This method transforms data to have a mean of 0 and a standard deviation of 1. It's particularly useful for algorithms that assume a Gaussian distribution or are sensitive to feature scales, such as SVMs and linear regression.

```python
from sklearn.preprocessing import StandardScaler

# Select numerical columns for scaling
numerical_cols = ['attendance', 'assignment', 'quizzes']

# Create a copy of the imputed DataFrame for standardization
df_standardized = df_imputed_mean.copy()

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the numerical columns
df_standardized[numerical_cols] = scaler.fit_transform(df_standardized[numerical_cols])

print("DataFrame after Standardization (Z-score normalization):")
display(df_standardized)
print("\nDescriptive statistics after Standardization:")
display(df_standardized[numerical_cols].describe())
```

### Handling Missing Values: Feature Scaling - Normalization (Min-Max scaling)

This method scales data to a fixed range, usually between 0 and 1. It's often used in neural networks and algorithms that require features to be within a specific range.

```python
from sklearn.preprocessing import MinMaxScaler

# Create a copy of the imputed DataFrame for normalization
df_normalized = df_imputed_mean.copy()

# Initialize the MinMaxScaler
min_max_scaler = MinMaxScaler()

# Fit and transform the numerical columns
df_normalized[numerical_cols] = min_max_scaler.fit_transform(df_normalized[numerical_cols])

print("DataFrame after Normalization (Min-Max scaling):")
display(df_normalized)
print("\nDescriptive statistics after Normalization:")
display(df_normalized[numerical_cols].describe())
```

You can choose either standardization or normalization based on the requirements of your machine learning model. Both methods ensure that all features contribute equally to the model's learning process.
